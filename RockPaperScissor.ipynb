```python
#importing libraries
from nis import match
import random

#how to play
print("Welcome to Rock paper scissors game")
print("press \n1 for Rock\n2 for Paper\n3 for Scissors")

arr = ["Rock", "Paper", "Scissors"]  

y= True
x = True
tries = 0   #to keep count of invalid tries
UserScore = 0  #to keep count of user score
CompScore = 0  #to keep count of computer score
mtch = 1   # to keep count of number of matches taking place

while(y):   # to keep the whole game running
    while(x):  # to make the user give valid input 
        try:
            choice = int(input("Enter choice"))   #edge case if the user enters decimal numbers or characters
        except:
            print("Choose between 1 and 3")
            tries += 1
        else:
            if(choice > 0 and choice < 4 ):
                print("You chose : " + arr[choice-1])
                x = False
                break
            else:
                print("you chose " +str(choice) + ", choose between 1 and 3")
                tries += 1
    
            if tries == 1:
                print("there was an input error")
            elif tries == 2:
                print("Maybe you typed in wrong.. again")
            elif tries == 3:
                print("Are you even trying")
            elif tries == 4:
                print("Lets just end the game")
                x = False
                exit
            
    print("Now it is computers turn")
    compChoice = random.randint(1,3)
    print("computer chose: " +arr[compChoice-1])

    if(choice == compChoice):
        print("----Its a tie-----")
    elif( (choice == 1 and compChoice == 2) or (choice == 2 and compChoice == 3) or (choice == 3 and compChoice == 1)):
        print("----Computer wins----")
        CompScore += 1
    elif((choice == 1 and compChoice == 3) or (choice == 2 and compChoice == 1) or (choice == 3 and compChoice == 2)):
        print("----You win----")
        UserScore += 1
        
    print("Your score is: " +str(UserScore) + "\nComputer score is : " + str(CompScore))
    
    decision = input("Do you want to play again?\n press 'y' to play again\npress 'n' to quit playing")
    Score = abs(CompScore - UserScore)
    
    if( decision == 'n' or decision == 'N'):
        print("Thank you for playing")
        if( CompScore > UserScore):
            print("The computer won by " + str(Score))
            y = False
            x = False
            exit
        elif(CompScore < UserScore):
            print("You won by" + str(Score))
            y = False
            x = False
            exit
        elif(CompScore == UserScore):
            print("its a tie")
            y = False
            x = False
            exit
                
    elif( decision == 'Y' or decision == 'y'):
        mtch += 1
        print("lets start match " + str(mtch))
    else:
        print("Invalid choice, closing the game")
        y= False
        x= False
        exit
            
```
